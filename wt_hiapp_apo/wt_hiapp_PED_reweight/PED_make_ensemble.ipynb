```python
from pathlib import Path


dir_path = Path('./')


pdb_path = dir_path / "hIAPP_disp_old.pdb"


with open(pdb_path, "r") as file:
    pdb_lines = file.readlines()


models = []
current_model = []
for line in pdb_lines:
    if line.startswith("MODEL"):
        if current_model:
            models.append(current_model)
        current_model = [line]
    else:
        current_model.append(line)
if current_model:
    models.append(current_model)

corrected_lines = []
for i, model in enumerate([models[0]] + models): 
    model_index = i + 1
    if model_index < 10:
        spacing = " " * 8
    elif model_index < 100:
        spacing = " " * 7
    elif model_index < 1000:
        spacing = " " * 6
    else:
        spacing = " " * 5
    corrected_lines.append(f"MODEL{spacing}{model_index}\n")
    for line in model[1:]:  
        corrected_lines.append(line)

corrected_file = pdb_path.with_name("hIAPP_disp.pdb")
with open(corrected_file, "w") as file:
    file.writelines(corrected_lines)

print(f"Saved: {corrected_file}")
```
