```python
# plot the timecourse 
def get_temps(nreps,tmax= 500, tmin= 300):
    """
    Compute the temperatures 
    """
    temps = []
    for i in range(0,nreps): 
        t=tmin*math.exp(i*math.log(tmax/tmin)/(nreps-1))
        temps.append(round(t,2))
    return temps
```

Compute rh values according to the following script info: https://www.biorxiv.org/content/10.1101/679373v2

```python
CA_atoms = trj.topology.select("name CA")
ca_trj = trj.atom_slice(CA_atoms, inplace=True)
rg = md.compute_rg(ca_trj)

Rg = rg * 10 

N = ca_trj.n_residues

def getRh(rg, N): 
    a1 = 0.216
    a2 = 4.06 
    a3 = 0.821

    return (rg) / ((a1*(rg-a2*N**(0.33)))/(N**(0.60)- N**(0.33))+a3)
Rh = getRh(Rg, N)/10

print("CA Hydrodynamic Radius:%6.3lf" %
      np.average(Rh), "+_%6.3lf" % block(Rh)**.5)
print("1st Half CA Hydrodynamic Radius:%6.3lf" % np.average(
    Rh[0:half]), "+_%6.3lf" % block(Rh[0:half])**.5)
print("2nd Half CA Hydrodynamic Radius:%6.3lf" % np.average(
    Rh[half:-1]), "+_%6.3lf" % block(Rh[half:-1])**.5)
```

```python
def plot_dihed(var1, var2, weights=None, title=" ", filename=" "):
    
    nbins = 40
    fig = plt.figure(figsize=(3, 3))

    v1 = var2
    v2 = var1

  
    if weights is not None:
   
        prob_reweighted, xedges, yedges = np.histogram2d(
            v1, v2, bins=nbins, 
            range=[[-3.14,3.14],[-3.14,3.14]], 
            density=False, weights=weights
        )
    else:
        
        prob_reweighted, xedges, yedges = np.histogram2d(
            v1, v2, bins=nbins, 
            range=[[-3.14,3.14],[-3.14,3.14]], 
            density=False
        )
    
  
    prob_reweighted /= np.sum(prob_reweighted)

    prob_reweighted = np.flipud(prob_reweighted)
    

    plt.rcParams["axes.edgecolor"] = "black"
    plt.rcParams["axes.linewidth"] = 1.0

 
    im = plt.imshow(prob_reweighted, interpolation='gaussian', extent=[yedges[0], yedges[-1], xedges[0], xedges[-1]], cmap=plt.cm.jet,  aspect='auto', vmin=0.0, vmax=0.10)
    plt.xlabel(r'$\phi$', labelpad=5)
    plt.ylabel(r"$\psi$", labelpad=5)
    plt.title(title, pad=20)
    
   
    cbar = plt.colorbar(im)
    cbar.ax.tick_params()
    cbar.set_label('Probability', labelpad=10, rotation=270)


    plt.tight_layout()
    filepath = os.path.join(outdir, filename)
    plt.savefig(filepath, dpi=400, bbox_inches="tight")
```

```python
# for the first frame make a matrix of shape # pairs, 7
# one-hot encode pairs of contacts
import tensorly as tl 
from tensorly.decomposition import partial_tucker
from tensorly.tenalg import multi_mode_dot

target_rank = [2, None, None]
core = None 
factors = None
batch_size = 200
total_slices = circuit_matrix.shape[0] // batch_size 

# for batch in data_generator(batch_size, total_slices): 

# chunk the incremental pca operations 
prev = 0 
current = batch_size
first_batch = circuit_matrix[prev:current,:]

core, factors = partial_tucker(first_batch, modes=[0, 1], rank=target_rank)
current +=batch_size

# fit the transformer 
for i in range(int(circuit_matrix.shape[0]/500)+1): 

    if i > circuit_matrix.shape[1]: 
        X = circuit_matrix[prev:,:]
        one_hot_matrix = np.zeros(shape=(X.shape[0], X.shape[1], 7))

        for k in range(1,8): 
            indices = np.argwhere(X == k)
            X_indices = indices[:,1]
            one_hot_matrix[:, X_indices, k-1] = 1

        projected_core = multi_mode_dot(one_hot_matrix, [tl.transpose(f) for f in factors], modes=[0, 1])
        core = (core + projected_core) / 2  # Weighted averaging (can be adjusted)
        core, factors = partial_tucker(tl.tucker_to_tensor((core, factors)), modes=[0, 1], rank=target_rank)

        break
    else: 
        X = circuit_matrix[prev:current,:]
        one_hot_matrix = np.zeros(shape=(X.shape[0], X.shape[1], 7))

        for k in range(1,8): 
            indices = np.argwhere(X == k)
            X_indices = indices[:,1]
            one_hot_matrix[:, X_indices, k-1] = 1

        projected_core = multi_mode_dot(one_hot_matrix, [tl.transpose(f) for f in factors], modes=[0, 1])
        core = (core + projected_core) / 2  # Weighted averaging (can be adjusted)
        core, factors = partial_tucker(tl.tucker_to_tensor((core, factors)), modes=[0, 1], rank=target_rank)
    current +=batch_size
```
